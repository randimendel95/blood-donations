# Import modules

```python
from sklearn.datasets import load_boston
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import scale
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from sklearn.cross_validation import KFold
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import pandas as pd
```

# Load data

```python
file = pd.read_csv('donation.csv', encoding='iso-8859-1')

donate = file['Made Donation in March 2007']
donate = np.array(donate)
data = file.drop('Made Donation in March 2007', axis=1)
data = np.array(data)
```

```python
print(data)
print(donate)
```

# Train-Test split

```python
data_train, data_test, don_train, don_test = train_test_split(data, donate, test_size=0.30, random_state=11)
```

```python
print(data_train.shape)
print(data_test.shape)
print(don_train.shape)
print(don_test.shape)
```

# Scale data

```python
mins = np.min(data_train, axis = 0)
maxes = np.max(data_train, axis = 0)
diff = maxes - mins
```

```python
diff
```

```python
data_train_stand = (data_train - mins) / diff
data_test_stand = (data_test - mins) / diff
```

```python
minDon = np.min(don_train)
maxDon = np.max(don_train)
diffDonate = maxDon - minDon
```

```python
don_train_stand = (don_train - minDon) / diffDonate
don_test_stand = (don_test - minDon) / diffDonate
```

# K-Fold CV

```python
kValues = [1, 2, 3, 4, 9, 10, 11, 12, 13]
```

```python
folds = KFold(len(data_train_stand), n_folds = 10, shuffle = True)
```

```python
scores = {}
for k in kValues:
    currentScores = []
    for train_index, val_index in folds:
        current_train_data, current_val_data = data_train_stand[train_index], data_train_stand[val_index]
        current_train_don, current_val_don = don_train_stand[train_index], don_train_stand[val_index]
        neigh = KNeighborsRegressor(n_neighbors = k)
        neigh.fit(current_train_data, current_train_don)
        guesses = neigh.predict(current_val_data)
        rmse = np.sqrt(mean_squared_error(guesses, current_val_don))
        currentScores.append(rmse)
    scores[k] = currentScores
```

# Plot Results

Plot your training accuracy across all folds as a function of K. What do you see?

```python
keys = sorted(scores.keys())
means = []
stdevs = []
for each in keys:
    current = scores[each]
    means.append(np.mean(current)) 
    stdevs.append(np.std(current))
    
```

```python
figure = plt.figure()
plt.plot(keys, means, 'bo:')
plt.xlabel('K')
plt.ylabel('Normalized RMSE')
plt.title('Error as a function of # of neighbors')
```

```python
figure = plt.figure()
plt.plot(keys, means, 'bo:')
plt.xlabel('K')
plt.ylabel('Normalized RMSE')
plt.title('Error as a function of # of neighbors')
plt.xlim([8, 13])
plt.ylim([.4,.42]);
```

Based on this chart, it appears that 9 is the most accurate value for k.
